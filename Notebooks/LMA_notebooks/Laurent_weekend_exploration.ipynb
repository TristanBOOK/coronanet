```python
def salut(pays, date, jours, Type='Deaths'):
    j14 = date+datetime.timedelta(days=jours)
    
    if j14 < pd.to_datetime('2020-12-04'):
        try:
            tmp = df_cases[(df_cases['ID']==pays) & (df_cases['Type']==Type)]
            return int(tmp[tmp['Date']==j14]['Cases'])
        except:
            return np.nan
```

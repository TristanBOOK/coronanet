# Deaths

```python
import pandas as pd
death_df=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
```

```python
death_df
```

```python
df = death_df[death_df['Province/State'].isnull()]
```

```python
df
```

```python
df = df.transpose()
df
```

```python
df.drop(['Lat','Long'], inplace = True)
```

```python
df.columns = df.iloc[1]
```

```python
df.drop(['Country/Region', 'Province/State'], inplace = True)
```

```python
ex_df = df[['France']]
```

```python
ex_df
```

## ARIMA

```python
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.stattools import adfuller
```

```python
ex_df.dtypes
```

```python
ex_df = ex_df.astype(int).dropna()
```

```python
zero_diff = ex_df.France
first_order_diff = ex_df.France.diff(1)
second_order_diff = ex_df.France.diff(1).diff(1)
```

```python
first_order_diff
```

```python
import matplotlib.pyplot as plt
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,4))
ax1.plot(zero_diff); ax1.set_title('Original Series')
ax2.plot(first_order_diff); ax2.set_title('1st Order Differencing')
ax3.plot(second_order_diff); ax3.set_title('2nd Order Differencing')
```

```python
print('p-value zero-diff: ', adfuller(ex_df.France)[1])
print('p-value first-diff: ', adfuller(ex_df.France.diff().dropna())[1])
print('p-value second-diff: ', adfuller(ex_df.France.diff().diff().dropna())[1])
```

## Lets try with a SARIMAX in that case

```python
smodel = pm.auto_arima(test.dropna(), seasonal=True, m=7, 
                       start_p=0, max_p=1, max_d=1, start_q=0, max_q=1,
                       start_P=0, max_P=2, max_D=1, start_Q=0, max_Q=2, 
                       trace=True, error_action='ignore', suppress_warnings=True)
```

## I'll try on the whole set

```python
pm.auto_arima(train, seasonal=True, m=52, 
                       start_p=0, max_p=2, max_d=2, start_q=0, max_q=2,
                       start_P=0, max_P=2, max_D=2, start_Q=0, max_Q=2, 
                       trace=True, error_action='ignore', suppress_warnings=True)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
sarima = SARIMAX(train, order=(0, 2, 1),seasonal_order=(0,0,0,52))
sarima= sarima.fit()
```

# What now ?

```python
sarima.summary()
```

```python
results = sarima.get_forecast(len(test), alpha=0.05)
forecast = results.predicted_mean
confidence_int = results.conf_int()
```

```python
confidence_int
```

```python
forecast
```

```python
plt.plot(forecast)
```

```python
central = pd.Series(forecast, index=test.index)
lower = pd.Series(confidence_int['lower France'], index=test.index)
upper = pd.Series(confidence_int['upper France'],  index =test.index)
```

```python
plt.figure(figsize=(12,5), dpi=200)
plt.plot(train, label = "train values")
plt.plot(test, label = "true test values")
plt.plot(central, color='darkgreen',label = "forecast")
plt.fill_between(lower.index, 
                 lower, 
                 upper, 
                 color= 'green', alpha=.15)

plt.title("SARIMA Forecast")
plt.legend()
plt.show()
```

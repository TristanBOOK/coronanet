## Preproc de Cases

```python
cases_df = cases_df2.copy()
cases_df = cases_df.drop(columns=['Age','Sex'])
cases_df.Date =  pd.to_datetime(cases_df.Date)
cases_df['Cases_New'] = cases_df['Cases_New'].astype('float')
cases_df['Cases'] = cases_df['Cases'].astype('float')
cases_df['Cases'] = pd.to_numeric(cases_df['Cases'], downcast='float')
cases_df['Cases_New'] = pd.to_numeric(cases_df['Cases_New'], downcast='float')
cases_df = cases_df[cases_df['Source']=='JHU']
cases_df = cases_df[cases_df.Type == 'Deaths']
```

# Merge des bases

```python
df_paul = df.copy()
df_death = deaths.copy()
df_cases = cases_df.copy()
```

```python
test_merge = df_paul.merge(df_cases, how='inner', left_on=['Date', 'ID'], right_on=['Date', 'ID'])
test_merge['Cases'] = test_merge['Cases'].astype('int32')
```

```python
test_merge.head()
```

## Ajout de features

```python
merge = test_merge.copy()
merge['J14'] = -10000
```

```python
for i in range(merge.shape[0]):
    j14 = merge['Date'][i]+datetime.timedelta(days=14)
    try:
        if j14 < pd.to_datetime('2020-12-04'):
            merge['J14'][i] = (df_cases[(df_cases['Date'] == (merge['Date'][i]+datetime.timedelta(days=14))) &
                        (df_cases['ID'] == merge['ID'][i]) & (df_cases['Type'] == 'Deaths')]['Cases'])
    except:
        np.nan
```

```python
merge[merge['J14'].isnull()][merge['ID']=='']
```

```python
def coucou(date, pays, Type='Deaths'):
    j14 = date + datetime.timedelta(days=14)
    if j14 < pd.to_datetime('2020-12-04'):
        tmp = df_cases[(df_cases['ID']==pays) & (df_cases['Type']==Type)]
        return tmp[tmp["Date"] == j14]['Cases']
```
